<h2>5. Faça um programa que peça as 4 notas bimestrais de um aluno e
mostre a média aritmética delas.  </h2>

```python
nota1 = float(input("Nota do primeiro bimestre: "))
nota2 = float(input("Nota do segundo bimestre: "))
nota3 = float(input("Nota do terceiro bimestre: "))
nota4 = float(input("Nota do quarto bimestre: "))
media = ((nota1 +nota2 + nota3 +nota4)/4)
print("A média é:", media)
```

<h2>7. Faça um programa que peça o raio de um círculo, calcule e mostre

```python
#Obs: Fórmula da área de um círculo: A = 3,14*r2, onde r é o raio. 
raio = float(input("Insira o raio do círculo: "))
area = (3.14*(raio**2))
print(area)
```

 </h2>

<h2> 9. Faça um programa que peça a temperatura em graus Fahrenheit (F),
transforme e mostre a temperatura em graus Celsius (°C). </h2>

```python
#°C = (5 * (F-32) / 9) 
# Fahrenheit para Celsius
temperatura_f = float(input("Insira a temperatura em Fahrenheit: "))
temperatura_c = (5*(temperatura_f - 32 )/9)
print(temperatura_f,"graus Fahrenheit são",temperatura_c,"graus Celsius")
```

```python
#°C = (5 * (F-32) / 9) 
# Celsius para Fahrenheit
temperatura_c = float(input("Insira a temperatura em Celsius: "))
temperatura_f = ((9/5*temperatura_c)+32)
print(temperatura_c,"graus Celsius são",temperatura_f,"graus Fahrenheit")
```

<h2>12. Faça um programa que peça o peso e altura de uma pessoa e calcule
seu IMC (Índice de Massa Corporal).
Obs: IMC = Peso/Altura2
 </h2>

```python
peso = float(input("Insira o seu peso: "))
altura = float(input("Insira a sua altura: "))
imc = (peso/(altura**2))
print(imc)
```
